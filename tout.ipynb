```python
from langchain.embeddings import HuggingFaceBgeEmbeddings

model_name = "BAAI/bge-small-en-v1.5"
encode_kwargs = {'normalize_embeddings': True} # set True to compute cosine similarity

bge_embeddings = HuggingFaceBgeEmbeddings(
    model_name=model_name,
    model_kwargs={'device': 'cpu'},
    encode_kwargs=encode_kwargs
)
```

#Dataset


```python
from datasets import load_dataset
from datasets import Dataset
import pandas as pd

# Load the Hugging Face dataset
dataset = pd.read_csv("/content/drive/MyDrive/IA Chrétien/collecte de données /bible_cleaned.csv")

# Convert to a Pandas DataFrame
df = dataset

# Preview the first few rows
df.head()
```

```python
from langchain.document_loaders import DataFrameLoader
import pandas as pd
docs = df['Text'].tolist()
```

```python
from langchain.document_loaders import DataFrameLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.docstore.document import Document # Import the Document class

# Assuming 'df' is your DataFrame
# Create documents using the Document class
docs = [Document(page_content=text) for text in df['Text'].tolist()]

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)

# Now split the documents
split_docs = text_splitter.split_documents(docs)
```

```python
print("Num Split Docs: ", len(split_docs))
```

```python
### Load to a database - Run this for the first time to create the Db file

# db = Chroma.from_documents(split_docs, bge_embeddings, persist_directory="./chroma_db")

## Load from a database
db = Chroma(persist_directory="./chroma_db", embedding_function=bge_embeddings)
```

```python
# question
question = "En combien de jour Dieu a créé la tere?"
matched_docs = db.similarity_search(question, k=8)

# print results
for index, value in enumerate(matched_docs):
    pos = index+1
    if index <=3:
        print(f"Matched doc {pos} is : ", matched_docs[index].page_content, "/n ===========")
```

```python
from langchain.prompts import PromptTemplate
prompt_template = """Contexte : Vous êtes un expert biblique assistant dans l’interprétation des Écritures. Vous avez accès aux versets suivants :
{retrieved_passages}

Informations supplémentaires : Les versets cités proviennent du livre de {Book Name}, chapitre {Chapter}, versets {Verse}. Ce passage aborde {sujet_du_passage}.

Question : {question}
Tâche : Fournissez une réponse claire et complète basée sur les versets cités. Votre réponse doit inclure une interprétation théologique et, si nécessaire, des références croisées avec d'autres passages pertinents.
:"""
PROMPT = PromptTemplate(
    template=prompt_template, input_variables=["context", "question"]
)
```

```python
from langchain.chains import RetrievalQA
from transformers import T5Tokenizer, T5ForConditionalGeneration
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

# Charger Flan-T5 pour la génération de texte
tokenizer = T5Tokenizer.from_pretrained('google/flan-t5-small')
model = T5ForConditionalGeneration.from_pretrained('google/flan-t5-small')

# Créer un pipeline de génération Hugging Face basé sur le modèle
hf_pipeline = pipeline("text2text-generation", model=model, tokenizer=tokenizer)

# Utiliser HuggingFacePipeline comme LLM pour Langchain
llm = HuggingFacePipeline(pipeline=hf_pipeline)

# Configurer le retrieval (en utilisant votre base de données locale, db)
retriever = db.as_retriever()
retriever.search_kwargs['k'] = 8

# Création de la chaîne RetrievalQA avec le modèle Flan-T5 et le retriever
chain_type_kwargs = {"prompt": PROMPT}  # Utiliser votre template de prompt ici
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever, chain_type_kwargs=chain_type_kwargs)

```

```python
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from transformers import T5Tokenizer, T5ForConditionalGeneration
from langchain.llms import HuggingFacePipeline
from transformers import pipeline

# Création du prompt avec le PromptTemplate
PROMPT = PromptTemplate(
    input_variables=["query"],
    template="""
    Question: {query}
    Réponse:
    """
)

# Récupération du modèle HuggingFace T5
tokenizer = T5Tokenizer.from_pretrained('t5-small')
model = T5ForConditionalGeneration.from_pretrained('t5-small')
hf_pipeline = pipeline("text2text-generation", model=model, tokenizer=tokenizer)

# Initialiser un LLM HuggingFace pipeline pour langchain
llm = HuggingFacePipeline(pipeline=hf_pipeline)

# Configuration de la chaîne RetrievalQA
chain_type_kwargs = {"prompt": PROMPT}

# Création de la chaîne RetrievalQA avec un modèle HuggingFace et le retriever
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever, chain_type_kwargs=chain_type_kwargs)

# Exemple de question
query = "Qui est Job dans la Bible ?"

# Invocation sans les champs supplémentaires
result = qa.invoke({
    "query": query
})

# Affichage de la réponse générée
print(f"Réponse : {result}")
```
