```python
def generate_embedding(question):
    inputs = tokenizer(question, return_tensors='pt')
    with torch.no_grad():
        outputs = model(**inputs)
        return outputs.pooler_output.numpy()
```

```python
def search_versets(question, k=5):  # k est le nombre de résultats à récupérer
    question_embedding = generate_embedding(question)
    distances, indices = index.search(np.array(question_embedding).astype('float32'), k)
    return indices, distances  # retourne les indices et distances des versets trouvés
```

```python
questions = [
    "Que dit la Bible sur l'amour ?",
    "Comment Dieu nous guide-t-il ?",
    "Quel est le message de la paix dans la Bible ?"
]

for question in questions:
    indices, distances = search_versets(question)
    print(f"Question: {question}")
    print("Versets récupérés:")
    for index in indices[0]:  # Indices des versets
        verse = df.iloc[index]  # Accéder à la ligne correspondante
        print(f"{verse['Book Name']} {verse['Chapter']}:{verse['Verse']} - {verse['Text']}")
    print("\n")
```

```python
# Exemple avec BERT
from transformers import BertTokenizer, BertModel
import torch

# Charger le tokenizer et le modèle BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Préparer la question
question = "quels sont les 10 commendements de Dieu"
inputs = tokenizer(question, return_tensors='pt', truncation=True, padding=True, max_length=512)

# Obtenir l'embedding de la question
with torch.no_grad():
    outputs = model(**inputs)
    question_embedding = torch.mean(outputs.last_hidden_state, dim=1).squeeze().numpy()
```

```python
#  Vérification de la forme de l'embedding
print("Shape of question_embedding before reshape:", question_embedding.shape)  # Devrait être (768,)

# Reshaper et vérifier le type
question_embedding = question_embedding.reshape(1, -1).astype(np.float32)  # Assurer que c'est (1, 768) et float32

# Vérification de la forme après le reshape
print("Shape of question_embedding after reshape:", question_embedding.shape)  # Devrait être (1, 768)

# Project the embedding to the correct dimension
question_embedding = np.concatenate((question_embedding, np.zeros((1, 4))), axis=1) # Add 4 columns of zeros to match the index dimension

# Effectuer la recherche
k = 5
D, I = index.search(question_embedding, k)  # D: distances, I: indices des voisins

# Afficher les résultats
print("Distances:", D)
print("Indices des voisins:", I)
```
